#### Carregar os datasets e transformar os dados

Limpa, transforma e grava as informações climáticas de 01/01/2010 à 30/04/2021 em uma tabela de SQLite 

```python
# conectar com o sqlite3 
# como o banco ainda não existe, esse comando irá criá-lo

conn = sqlite3.connect('C:/tcc_ds_puc/db_sqlite/dataset.db')
cur = conn.cursor() 
```

```python
# Lê os arquivos de clima no diretório e grava em uma lista

files = []
path = 'C:/tcc_ds_puc/raw/clima/'
processados = 'C:/tcc_ds_puc/raw/clima/processados/'
for (dirpath, dirnames, filenames) in walk(path):
    files.extend(filenames)
    break

# seleciona apenas os arquivos ".csv" e grava na lista
    
files_csv = []
for i in files:
    if '.csv' in i:
        files_csv.append(i)
    else:
        pass

print('Total de arquivos ".csv":',len(files_csv),'\n')
```

```python
%%time
# Lê os arquivos da lista, faz a limpeza dos dados e grava em uma tabela no banco dataset.db

# Nome das colunas

columns = [
           'DT_MEDICAO','HORA_MEDICAO_UTC','PRECIPITACAO_TOTAL_HORARIO_MM',
           'PRESSAO_ATMOSFERICA_AO_NIVEL_DA_ESTACAO_HORARIA_MB',
           'PRESSAO_ATMOSFERICA_REDUZIDA_NIVEL_MAR+AUT_MB', 
           'PRESSAO_ATMOSFERICA_MAX_NA_HORA_ANT__AUT_MB',
           'PRESSAO_ATMOSFERICA_MIN_NA_HORA_ANT_AUT_MB',
           'RADIACAO_GLOBAL_KJM2',
           'TEMPERATURA_CPU_ESTACAO',
           'TEMPERATURA_DO_AR_BULBO_SECO_HORARIA_C',
           'TEMPERATURA_DO_PONTO_DE_ORVALHO_C',
           'TEMPERATURA_MAXIMA_NA_HORA_ANT_AUT_C',
           'TEMPERATURA_MINIMA_NA_HORA_ANT_AUT_C',
           'TEMPERATURA_ORVALHO_MAX_NA_HORA_ANT_AUT_C',
           'TEMPERATURA_ORVALHO_MIN_NA_HORA_ANT_AUT_C',
           'TENSAO_BATERIA_ESTACAO',
           'UMIDADE_REL_MAX_NA_HORA_ANT_AUT_PERC',
           'UMIDADE_REL_MIN_NA_HORA_ANT_AUT_PERC',
           'UMIDADE_RELATIVA_DO_AR_HORARIA_PERC',
           'VENTO_DIREÇAO_HORARIA_GR',
           'VENTO_RAJADA_MAXIMA_MS',
           'VENTO_VELOCIDADE_HORARIA_MS',
           'ESTACAO',
           'COD_ESTACAO',
           'LATITUDE',
           'LONGITUDE',
           'ALTITUDE',
          ]

# tratamento e transformação dos dados

for file in files_csv:
    df = pd.read_csv(path+file,skiprows=11,sep=";",header=None)
    df.iloc[:,1] = ('000'+df.iloc[:,1].astype('str')).str[:-2]
    header = pd.read_csv(path+file,nrows=9,header=None,sep=";",encoding="ISO-8859-1")
    header = header.values.tolist()
    df = df.iloc[:,0:22]
    dicionario = {}
    for i in header:
        i = str(i).strip('[]').strip("''")
        key = (i[0:i.index(':')])
        value = (i[i.index(':')+1:])
        dicionario[key] = value
        head = pd.DataFrame([dicionario])

    estacao = head.iloc[:,0].loc[0][1:]
    codigo = head.iloc[:,1].loc[0][1:]
    latitude = head.iloc[:,2].loc[0][1:]
    longitude = head.iloc[:,3].loc[0][1:]
    altitude = head.iloc[:,4].loc[0][1:].replace('F','0')
    df.describe()
    
    df['ESTACAO'] = estacao
    df['COD_ESTACAO'] = codigo 
    df['LATITUDE'] = latitude  
    df['LONGITUDE'] = longitude
    df['ALTITUDE'] = altitude
    df.replace({':': ''}, regex=True, inplace=True)
    df.replace({',': '.'}, regex=True, inplace=True)
    df.replace({'null': ''}, regex=True, inplace=True)
    df.columns = columns
    #df['DT_MEDICAO']
    #df['DT_MEDICAO'] = pd.to_datetime(df["DT_MEDICAO"]).dt.date
    
    df = df.astype({'HORA_MEDICAO_UTC': int,
                    'PRECIPITACAO_TOTAL_HORARIO_MM': float,
                    'PRESSAO_ATMOSFERICA_AO_NIVEL_DA_ESTACAO_HORARIA_MB': float,
                    'PRESSAO_ATMOSFERICA_MAX_NA_HORA_ANT__AUT_MB': float,
                    'PRESSAO_ATMOSFERICA_MIN_NA_HORA_ANT_AUT_MB': float,
                    'RADIACAO_GLOBAL_KJM2': float,
                    'TEMPERATURA_DO_AR_BULBO_SECO_HORARIA_C': float,
                    'TEMPERATURA_DO_PONTO_DE_ORVALHO_C': float,
                    'TEMPERATURA_MAXIMA_NA_HORA_ANT_AUT_C': float,
                    'TEMPERATURA_MINIMA_NA_HORA_ANT_AUT_C': float,
                    'TEMPERATURA_ORVALHO_MAX_NA_HORA_ANT_AUT_C': float,
                    'TEMPERATURA_ORVALHO_MIN_NA_HORA_ANT_AUT_C': float,
                    'UMIDADE_REL_MAX_NA_HORA_ANT_AUT_PERC': float,
                    'UMIDADE_REL_MIN_NA_HORA_ANT_AUT_PERC': float,
                    'UMIDADE_RELATIVA_DO_AR_HORARIA_PERC': float,
                    'VENTO_DIREÇAO_HORARIA_GR': float,
                    'VENTO_RAJADA_MAXIMA_MS': float,
                    'VENTO_VELOCIDADE_HORARIA_MS': float,
                    'LATITUDE': float,
                    'LONGITUDE': float,
                    'ALTITUDE': float})
    
    # transforma os valores negativos em nulos
    df.where(df != -9999.0, np.nan, inplace=True)
   
    # grava os dados do DataFrame na tabela do SQLite
    df.to_sql('CLIMA_INMET',
              conn,
              if_exists='append',
              index=False)
    
    # Após a gravação dos dados na tabela, move os arquivo para uma pasta de arquivos processados
    shutil.move(path+file, processados+file)
```

```python
%%time
# cria o índice da tabela

index = """
CREATE INDEX index_CLIMA_INMET
    ON CLIMA_INMET(
                    DT_MEDICAO
                   ,HORA_MEDICAO_UTC
                   ,ESTACAO
                   ,COD_ESTACAO
                   )
"""
cur.execute(index)
conn.commit()

print('A tabela está pronta para uso!')
```

Transforma e grava os registros de focos de 01/01/2010 à 30/04/2021 em uma tabela de SQLite

```python
%%time
# grava o nome dos arquivos do diretório na lista chamada "files"

files = []
path = 'C:/tcc_ds_puc/raw/focos/'
processados = 'C:/tcc_ds_puc/raw/focos/processados/'
for (dirpath, dirnames, filenames) in walk(path):
    files.extend(filenames)
    
# seleciona apenas os arquivos .csv e grava na lista files_csv
    
files_csv = []
for i in files:
    if '.csv' in i:
        files_csv.append(i)
    else:
        pass

# apresenta na tela a quantidade de arquivos que vamos utilizar

print('Total de arquivos ".csv":',len(files_csv),'\n')
```

```python
%%time

# Lê os arquivos, transforma, faz a limpeza dos dados e grava em uma tabela no banco dataset.db

# grava o nome das colunas na variável columns
columns = ['DATA_HORA','SATELITE','PAIS','ESTADO','MUNICIPIO',
           'BIOMA','DIASEMCHUVA','PRECIPITACAO','RISCO_FOGO',
           'LATITUDE','LONGITUDE','FRP']

# tratamento e transformação dos dados
for file in files_csv:
    df = pd.read_csv(path+file,skiprows=1,sep=",",header=None)
    df.replace({',': '.'}, regex=True, inplace=True)
    df.columns = columns
    df['DT_FOCO'] = df.iloc[:,0].str[0:10].replace('/','-')
    df['DT_FOCO'] = pd.to_datetime(df['DT_FOCO']).dt.date
    df['HR_FOCO'] = df.iloc[:,0].str[11:13]
    df['DT_FOCO'] = pd.to_datetime(df["DT_FOCO"]).dt.date
    df.replace({np.nan: -999},inplace=True)
    df = df.astype({'HR_FOCO': int,
                    'DIASEMCHUVA': int,
                    'PRECIPITACAO': float,
                    'RISCO_FOGO': float,
                    'LATITUDE': float,
                    'LONGITUDE': float,
                    'FRP': float})
    
    # transforma os valores negativos em nulos
    df.where(df != -999, np.nan, inplace=True)
    
    # grava os dados do DataFrame na tabela do SQLite
    df.to_sql('FOCOS_INPE',
              conn,
              if_exists='append',
              index=False)
    
    # Após a gravação dos dados na tabela, move os arquivo para uma pasta de arquivos processados
    shutil.move(path+file, processados+file)
```

```python
%%time

# cria o indice da tabela FOCOS_INPE

index = """
CREATE INDEX index_FOCOS_INPE
    ON FOCOS_INPE(
                   DT_FOCO
                  ,HR_FOCO
                  ,ESTADO
                  ,MUNICIPIO
                  ,BIOMA
                 )
"""
cur.execute(index)
conn.commit()

print('A tabela está pronta para uso!')
```

#### Cria as tabelas com os dados limpos e as variáveis modeladas

```python
# cria uma tabela com as variáveis que vamos utilizar sobre os focos de incêndio 

varfocos_create = """
CREATE TABLE VAR_FOCOS_INPE AS

SELECT 
     DT_FOCO
    ,ESTADO
    ,MUNICIPIO
    ,BIOMA
    ,LATITUDE
    ,LONGITUDE
    ,1 AS FOCO_REGISTRADO
FROM FOCOS_INPE
GROUP BY 
     DT_FOCO
    ,ESTADO
    ,MUNICIPIO
    ,BIOMA
    ,LATITUDE
    ,LONGITUDE

"""
varfocos_index = """
CREATE INDEX index_VAR_FOCOS_INPE
ON VAR_FOCOS_INPE
    (
    DT_FOCO,
    ESTADO,
    MUNICIPIO,
    BIOMA
    )
"""

cur.execute(varfocos_create)
cur.execute(varfocos_index)
conn.commit()
```

```python
# cria uma tabela com as variáveis modeladas para os registros de clima

varclima_create = """
CREATE TABLE VAR_CLIMA_INMET AS 
SELECT 
     DT_MEDICAO
    ,COD_ESTACAO
    ,ESTACAO
    ,LATITUDE
    ,LONGITUDE
    ,MAX(PRECIPITACAO_TOTAL_HORARIO_MM)+MIN(PRECIPITACAO_TOTAL_HORARIO_MM) AS PRECIPITACAO_IND
    ,MAX(PRESSAO_ATMOSFERICA_AO_NIVEL_DA_ESTACAO_HORARIA_MB)+MIN(PRESSAO_ATMOSFERICA_AO_NIVEL_DA_ESTACAO_HORARIA_MB) AS PRESSAO_ATM_EST_IND
    ,MAX("PRESSAO_ATMOSFERICA_REDUZIDA_NIVEL_MAR+AUT_MB")+MIN("PRESSAO_ATMOSFERICA_REDUZIDA_NIVEL_MAR+AUT_MB") AS PRESSAO_ATM_MAR_IND
    ,MAX(RADIACAO_GLOBAL_KJM2)+MIN(RADIACAO_GLOBAL_KJM2) AS RADIACAO_GLOBAL_IND
    ,MAX(TEMPERATURA_DO_AR_BULBO_SECO_HORARIA_C)+MIN(TEMPERATURA_DO_AR_BULBO_SECO_HORARIA_C) AS TEMPERATURA_DO_AR_BULBO_SECO_IND
    ,MAX(TEMPERATURA_DO_PONTO_DE_ORVALHO_C)+MIN(TEMPERATURA_DO_PONTO_DE_ORVALHO_C) AS TEMPERATURA_DO_PONTO_DE_ORVALHO_IND
    ,MAX(TEMPERATURA_MAXIMA_NA_HORA_ANT_AUT_C)+MIN(TEMPERATURA_MINIMA_NA_HORA_ANT_AUT_C) AS TEMPERATURA_IND
    ,MAX(UMIDADE_RELATIVA_DO_AR_HORARIA_PERC)+MIN(UMIDADE_RELATIVA_DO_AR_HORARIA_PERC) AS UMIDADE_RELATIVA_IND
    ,MAX(VENTO_RAJADA_MAXIMA_MS)+MIN(VENTO_RAJADA_MAXIMA_MS) AS VENTO_RAJADA_IND
    ,MAX(VENTO_VELOCIDADE_HORARIA_MS)+MIN(VENTO_VELOCIDADE_HORARIA_MS) AS VENTO_VELOCIDADE_IND
FROM CLIMA_INMET 
GROUP BY 
     DT_MEDICAO
    ,COD_ESTACAO
    ,ESTACAO
    ,LATITUDE
    ,LONGITUDE

"""
varclima_index = """
CREATE INDEX index_VAR_CLIMA_INMET
ON VAR_CLIMA_INMET
    (
     DT_MEDICAO
    ,COD_ESTACAO
    ,ESTACAO
    ,LATITUDE
    ,LONGITUDE
    )
"""

cur.execute(varclima_create)
cur.execute(varclima_index)
conn.commit()
```

Para aproveitar melhor os dados da tabela CLIMA_INMET, somei o registro de menor e maior valor capturado em cada variável por dia, 
dessa forma a quantidade de registros é reduzida a um por dia expresso por um único valor que é a soma do mínimo com o máximo.

Uma caracteristica desse dataset, já citada anteriormente é que só existem registros em datas que foram identificados focos de incêndio, então para uma análise adequada da série serão incluidos os dias em que nenhum foco foi registrado.

```python
# Função para gerar datas para o calendário

def gerar_data(ano, mes, dia):
    data = datetime.datetime(ano, mes, dia)
    return data

# Data Inicial
datas = datetime.datetime(2010, 1, 1)

# Somar 1 dia
d1 = datetime.timedelta(days=+1)

# Lista com as datas
lista_datas = []

# Data final
while datas <= datetime.datetime(2021, 4, 30):
    data_gerada = gerar_data(datas.year, datas.month, datas.day)
    data_no_formato = data_gerada.strftime('%Y-%m-%d')

    lista_datas.append(data_no_formato)

    datas = datas + d1

# cria um DataFrame semelhante ao "df" (DataFrame com os registros de focos) 
# contendo todos os dias no período de 01/01/2010 à 30/04/2021 com valor zero 
# para todas as variáveis.
    
df_calendario = pd.DataFrame(lista_datas,columns=['DT_FOCO'])
df_calendario['LATITUDE'] = 0.0
df_calendario['LONGITUDE'] = 0.0
df_calendario['FOCO'] = 0
```

```python
# Vizualização do DataFrame df_calendario

df_calendario.info()
```

```python
# Vizualização do DataFrame df

df.info()
```

```python
# Empilha os dois DataFrames e cria as colunas Dia, Mês e Ano.

df_total = df.append(df_calendario).reset_index(drop=True)
df_total['ANO'] = pd.to_datetime(df_total['DT_FOCO']).dt.year
df_total['MES'] = pd.to_datetime(df_total['DT_FOCO']).dt.month
df_total['DIA'] = pd.to_datetime(df_total['DT_FOCO']).dt.day
```

```python
# Reorganiza as colunas do DataFrame

colunas = ['DT_FOCO','ANO','MES','DIA','FOCO','LATITUDE','LONGITUDE']
df_total = df_total[colunas]
```

```python
# Vizualização do DataFrame df_total

print('Visualização do df_total:\n')

df_total
```

```python
# Resume o DataFrame com a soma dos focos por dia

df_focos = pd.DataFrame(df_total.groupby(['DT_FOCO','ANO','MES','DIA'])['FOCO'].sum()).reset_index()
```

```python
# Altera o tipo de dado do campo DT_FOCO (anteriormente esse campo era do tipo object)

df_focos['DT_FOCO'] = pd.to_datetime(df_focos['DT_FOCO'], format = '%Y-%m-%d')

# Vizualização do DataFrame df_focos

df_focos.info()
```

```python
# Gráfico com a quandidade de focos por dia na série total

sns.set_style("whitegrid")
fig, ax = plt.subplots()
ax.plot('DT_FOCO','FOCO',data=df_focos,) #color='red')
grade_ano =  mdates.MonthLocator(interval=12)
ax.xaxis.set_major_locator(grade_ano)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
datemin = np.datetime64(df_focos['DT_FOCO'][0], 'Y') + np.timedelta64(-1, 'Y')
datemax = np.datetime64(df_focos['DT_FOCO'][4137], 'Y') + np.timedelta64(1, 'Y')
ax.set_xlim(datemin, datemax)
ax.format_xdata = mdates.DateFormatter('%Y')
plt.title("Evolução da Quantidade de Focos", loc='left', fontsize=14, fontweight=0, color='black')
plt.xlabel("Ano")
plt.ylabel("Qt Focos",fontsize=12)
plt.xticks(rotation=30,fontsize=10);
```

O gráfico acima mostra a quantidade de focos registrados dia a dia no período de 01/01/2010 à 30/04/2021. Isso facilitará a identificação dos "endereços" com maiores registros.

```python
# Imports para trabalhar com polígonos e h3

from h3 import h3                             
import pyproj                                 
import shapely
from shapely import wkt                        
from shapely.geometry import Point, Polygon   
from shapely.ops import nearest_points        
from shapely.wkt import loads as load_wkt        
import matplotlib.pyplot as plt               
import mapclassify as mc
import libpysal.weights as weights, contextily

print("Versões das bibliotécas:\n")

%reload_ext watermark
%watermark --iversions
```

```python
# média da área em km2 do hexâgono criado
# https://h3geo.org/docs/core-library/restable/

h3_level = 6 # Average Hexagon Area (km2) 36.1290521

# função para atribuir o endereço h3 para cada latitude e longitude

def lat_lng_to_h3(row):
    return h3.geo_to_h3(row.LATITUDE, 
                        row.LONGITUDE, 
                        h3_level)

# aplicar a função para atribuir os endereços h3 no GeoDataFrame com os focos

gdf_focos['h3_focos'] = gdf_focos.apply(lat_lng_to_h3, axis=1)
```

```python
# Criar um novo DF com os endereços

df_focos_h3 = pd.DataFrame(gdf_focos)
del df_focos_h3['geometry']
```

```python
# Grava os registos na tabela 'FOCOS_H3' do SQLite

df_focos_h3.to_sql('FOCOS_H3',
                   conn,
                   if_exists='replace',
                   index=False)
```

```python
# Visualização do DataFrame com os endereços

print('Visualização do gdf_focos com os endereços mapeados:\n')

gdf_focos.head()
```

```python
# contar os focos em cada área h3 mapeada

gdf_focos['ANO'] = pd.to_datetime(gdf_focos['DT_FOCO']).dt.year
gdf_focos_count = gdf_focos.groupby(['h3_focos','ANO']).h3_focos\
                    .agg('count')\
                    .to_frame('count_focos')\
                    .reset_index()
```

```python
# cria uma função para atribuir um polígono do tamanho de cada área

def add_geometry(row):
    points = h3.h3_to_geo_boundary(
      row['h3_focos'], True)
    return Polygon(points)

# apolica a função no GeoDataFrame criando a coluna geometry

gdf_focos_count['geometry'] = gdf_focos_count.apply(add_geometry, axis=1)
```

```python
# adiciona um sistema de cordenadas em metros

gdf_focos_count = gpd.GeoDataFrame(gdf_focos_count, crs='EPSG:3395')
```

```python
# Cria um DataFrame vazio onde serão imputados os endereços

h3_pantanal_new = pd.DataFrame([],columns=['h3_pantanal',
                                           'geometry',
                                           'h3_centroid_pantanal',
                                           'lat_h3','lon_h3'])
```

```python
# Encontra o id h3 para cada polígono

h3_pantanal_old = h3.polyfill_geojson(gjs_pantanal, 6)
```

```python
# Usa um laço para extrair os atributos dos endereços

for h3_hexagono in h3_pantanal_old:    
    # cria uma variável com as bordas do polígono 
    geometry = shapely.geometry.Polygon(
        h3.h3_to_geo_boundary(h3_hexagono,geo_json=True))
    
    # cria uma variável com a ponto central do polígono 
    h3_centroid = shapely.geometry.Polygon(
        h3.h3_to_geo_boundary(h3_hexagono,geo_json=True)).centroid.wkt
    lat_lon = shapely.geometry.Polygon(
        h3.h3_to_geo_boundary(h3_hexagono,geo_json=True)).centroid
    
    # insere os endereços e seus atributos (borda e centróide) no DataFrame
    h3_pantanal_new.loc[len(h3_pantanal_new)]=[
        h3_hexagono,
        geometry,
        h3_centroid,
        lat_lon.y,
        lat_lon.x
        ]    

# Exibe os primeiros registros
h3_pantanal_new.head(2)
```

#### Tratamento das variáveis para a criação do modelo de Machine Learning

```python
# query para selecionar e tratar os dados que serão utilizados no modelo 

df = pd.read_sql_query("""

SELECT 
     A.MES_MEDICAO
    ,A.DT_MEDICAO
    ,A.PRECIPITACAO_IND
    ,A.RADIACAO_GLOBAL_IND
    ,A.TEMPERATURA_DO_AR_BULBO_SECO_IND
    ,A.TEMPERATURA_DO_PONTO_DE_ORVALHO_IND
    ,A.TEMPERATURA_IND
    ,A.UMIDADE_RELATIVA_IND
    ,A.VENTO_RAJADA_IND
    ,A.VENTO_VELOCIDADE_IND
    ,SUM(A.QT_FOCOS) OVER (ORDER BY A.MES_MEDICAO ROWS BETWEEN 6 PRECEDING AND 1 PRECEDING) AS FREQUENCIA
    ,A.FOCOS_IND
FROM (
SELECT 
     cast(strftime('%m', A.DT_MEDICAO) as float) AS MES_MEDICAO
    ,A.DT_MEDICAO
    ,A.PRECIPITACAO_IND
    ,A.RADIACAO_GLOBAL_IND
    ,A.TEMPERATURA_DO_AR_BULBO_SECO_IND
    ,A.TEMPERATURA_DO_PONTO_DE_ORVALHO_IND
    ,A.TEMPERATURA_IND
    ,A.UMIDADE_RELATIVA_IND
    ,A.VENTO_RAJADA_IND
    ,A.VENTO_VELOCIDADE_IND
    ,SUM(B.QT_FOCOS) AS QT_FOCOS
    ,CASE WHEN SUM(B.QT_FOCOS) >= 1 THEN 1 ELSE 0 END FOCOS_IND
FROM VAR_CLIMA_INMET AS A
    LEFT JOIN 
    (
    SELECT
         A.DT_FOCO
        ,SUM(A.FOCO_REGISTRADO) AS QT_FOCOS
    FROM VAR_FOCOS_INPE AS A
    GROUP BY
    A.DT_FOCO
    ) AS B
    ON  A.DT_MEDICAO = B.DT_FOCO
GROUP BY 
     A.DT_MEDICAO
    ,A.PRECIPITACAO_IND
    ,A.RADIACAO_GLOBAL_IND
    ,A.TEMPERATURA_DO_AR_BULBO_SECO_IND
    ,A.TEMPERATURA_DO_PONTO_DE_ORVALHO_IND
    ,A.TEMPERATURA_IND
    ,A.UMIDADE_RELATIVA_IND
    ,A.VENTO_RAJADA_IND
    ,A.VENTO_VELOCIDADE_IND
) AS A
              
""",conn)
```

A query acima junta os registros de focos de incêndio da tabela VAR_FOCOS_INPE com os registros da estação climática. Como foi utilizada apenas uma estação climática para trazer as variáveis climáticas para um território extenso, as coordenadas dos focos não serão aproveitadas no modelo. Além disso, para identificar o comportamento ao longo do tempo, foi criada a variável FREQUENCIA que aponta quantos focos foram registrados nos últimos 6 dias até o dia anterior aos registros das variáveis climáticas.

```python
# Resumo do DataFrame que será utilizado no modelo
# quantidade de linas e colunas

dim = str(df.shape)
lin = '{0:,}'.format(int(str([dim[1:dim.index(',')]])\
                         .strip('[]').strip('()')\
                         .replace("'",''))).replace(',','.')
col = '{0:,}'.format(int(str([dim[dim.index(',')+1:]])\
                         .strip('[]').replace(')','')\
                         .replace("'",''))).replace(',','.')

print('\nResumo do DataFrame original:')
print('  * Total de linhas: ',lin)
print('  * Total de colunas:   ',col)
```

```python
# Resumo das correlações por variável com a incidência de focos de incêndio

cor_target = abs(cor["FOCOS_IND"])
variaveis = cor_target[cor_target>0.0]

print('Resumo das correlações:\n')
variaveis.sort_values(ascending=False)
```

A única variável que apresentou correlação moderada com os focos registrados foi a UMIDADE_RELATIVA_IND com 53% de correlação. Todas as demais apresentaram fraca ou nenhuma correlação com os focos de incêndio que é a variável prevista.
Nesse estudo usar variáveis com baixa correlação pode não ter tanto impacto pois serão aplicados modelos de classificação.

```python
print('Definindo as variáveis preditoras:')

atributos = [
             'UMIDADE_RELATIVA_IND',
             'FREQUENCIA',
             'RADIACAO_GLOBAL_IND',
             'TEMPERATURA_DO_AR_BULBO_SECO_IND',
             'MES_MEDICAO',
             'PRECIPITACAO_IND',
              ]
print(atributos)
print('\nDefinindo a variável prevista:')
atrib_prev = ['FOCOS_IND']
print(atrib_prev)
```

```python
print('Renomeando variáveis preditoras para "x" e prevista para "y"')
x = df[atributos].values
y = df[atrib_prev].values
```

```python
print('Separando o dataset em treino e teste:')
split_teste_size = 0.30
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y,\
                                                        test_size = split_teste_size,\
                                                        random_state = 42)
print("{0:0.2f}% nos dados de treino".\
      format((len(x_treino)/len(df.index))*100))
print("{0:0.2f}% nos dados de teste".\
      format((len(x_teste)/len(df.index))*100))
```

```python
print('Resumo dataset original:')
print("Original True : {0} ({1:0.2f}%)".format(len(df.loc[df["FOCOS_IND"] == 1]),
                                                (len(df.loc[df["FOCOS_IND"] == 1])/len(df.index)*100)))
print("Original False : {0} ({1:0.2f}%)".format(len(df.loc[df["FOCOS_IND"] == 0]),
                                                (len(df.loc[df["FOCOS_IND"] == 0])/len(df.index)*100)))
print("")
print('Resumo base de treino:')
print("Training True : {0} ({1:0.2f}%)".format(len(y_treino[y_treino[:] == 1]),
                                                (len(y_treino[y_treino[:] == 1])/len(y_treino)*100)))
print("Training False : {0} ({1:0.2f}%)".format(len(y_treino[y_treino[:] == 0]),
                                                (len(y_treino[y_treino[:] == 0])/len(y_treino)*100)))
print("")
print('Resumo base de teste:')
print("Test True : {0} ({1:0.2f}%)".format(len(y_teste[y_teste[:] == 1]),
                                                (len(y_teste[y_teste[:]  == 1])/len(y_teste)*100)))
print("Test False : {0} ({1:0.2f}%)".format(len(y_teste[y_teste[:] == 0]),
                                                (len(y_teste[y_teste[:]  == 0])/len(y_teste)*100)))
```

Com o resumo acima, é possível notar que esses dados são balanceados, então não será necessário aplicar outra técnica de undersample ou oversample para ajustar os dados de treino.

```python
# Apresentando no gráfico a distribuição de verdadeiros e falsos no dataset

target = df.FOCOS_IND.value_counts()
print('False 0:', target[0])
print('True 1:', target[1])
print('Proportion:', round(target[0] / target[1], 2), ': 1')

target.plot(kind='bar',
            title='Distrubuição da base de treino',
            color = ['#1F77B4', '#FF7F0E']);
```

#### Treinamento e teste dos modelos de Machine Learning

```python
%%time
print('Modelo 1 - Teste com Naive Bayes\n')
modelo_v1 = GaussianNB()
modelo_v1.fit(x_treino, y_treino.ravel())
nb_predict_train = modelo_v1.predict(x_treino)
print("Acuracidade Treino: {0:4f}".format(metrics.accuracy_score(y_treino, nb_predict_train)))
nb_predict_test = modelo_v1.predict(x_teste)
print("Acuracidade Teste: {0:4f}".format(metrics.accuracy_score(y_teste, nb_predict_test)))

print('\nConfusion Matrix')
print("{0}".format(metrics.confusion_matrix(y_teste, nb_predict_test, labels = [1,0])))
print('')
print('Classification Report')
print(metrics.classification_report(y_teste, nb_predict_test, labels = [1,0]))
```

```python
print('Modelo 2 - Teste com Random Forest\n')
modelo_v2 = RandomForestClassifier(n_estimators=190, min_samples_leaf=16, random_state=42, max_depth=13)
modelo_v2.fit(x_treino, y_treino.ravel())
rf_predict_train = modelo_v2.predict(x_treino)
print("Acuracidade Treino: {0:4f}".format(metrics.accuracy_score(y_treino, rf_predict_train)))
rf_predict_test = modelo_v2.predict(x_teste)
print("Acuracidade Teste: {0:4f}".format(metrics.accuracy_score(y_teste, rf_predict_test)))

print('\nConfusion Matrix')
print("{0}".format(metrics.confusion_matrix(y_teste, rf_predict_test, labels = [1,0])))
print('')
print('Classification Report')
print(metrics.classification_report(y_teste, rf_predict_test, labels = [1,0]))
```

```python
print('Modelo 3 - Teste com Regressão Logística\n')
modelo_v3 = LogisticRegression(C = 0.7, random_state=42, max_iter=150)
modelo_v3.fit(x_treino, y_treino.ravel())
lr_predict_test = modelo_v3.predict(x_treino)
print("Acuracidade Treino: {0:4f}".format(metrics.accuracy_score(y_treino, lr_predict_test)))
lr_predict_test = modelo_v3.predict(x_teste)
print("Acuracidade Teste: {0:4f}".format(metrics.accuracy_score(y_teste, lr_predict_test)))
print('\nConfusion Matrix')
print("{0}".format(metrics.confusion_matrix(y_teste, lr_predict_test, labels = [1,0])))
print('')
print('')
print('Classification Report')
print(metrics.classification_report(y_teste, lr_predict_test, labels = [1,0]))
```

```python
print('Modelo 4 - Teste com XGBClassifier\n')
modelo_v4 = XGBClassifier(n_estimators=200, random_state=44, max_depth=14,
                          use_label_encoder=False, eval_metric='logloss')
modelo_v4.fit(x_treino, y_treino)
xgbc_predict_train = modelo_v4.predict(x_treino)
print("Exatidão (Accuracy): {0:4f}".format(metrics.accuracy_score(y_treino, xgbc_predict_train)))
xgbc_predict_test = modelo_v4.predict(x_teste)
print("Exatidão (Accuracy): {0:4f}".format(metrics.accuracy_score(y_teste, xgbc_predict_test)))
print('\nConfusion Matrix')
print("{0}".format(metrics.confusion_matrix(y_teste, xgbc_predict_test, labels = [1,0])))
print('')
print('Classification Report')
print(metrics.classification_report(y_teste, xgbc_predict_test, labels = [1,0]))
```

Dentre os modelos criados, o modelo que possui melhor acuracidade para o nosso estudo é o Modelo 2 - Randon Forest que apontou uma acuracidade de 77%.
